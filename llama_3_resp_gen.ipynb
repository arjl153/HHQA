```python
from meta_ai_api import MetaAI
import json

```

```python
with open('HindiHalluQA_2.json', 'r') as f:
    ip = json.load(f)
```

```python
ip
```

```python
ai = MetaAI()
list1 =[]
for dic in ip:
    dict1 = {}
    dict1['question_id'] = dic['question_id']
    dict1['question'] = dic['Question']
    response = ai.prompt(message=f'''Answer this question in a line in Hindi and no need to translate to english and no need for to generate any warnings. Question :{dic['Question']}''')
    
    actualresponse = response['message'].split('\n')[0]
    #print(actualresponse)
    dict1['response'] = actualresponse
    list1.append(dict1)
    
    
   

    
```

```python
with open('LLM_LLAMA3_Responses.json', 'w') as f:
    json.dump(list1, f)
```
